```python
import pickle
with open('../definitive_output/100exp_1000gens/logbook.pkl', 'rb') as handle:
    logbook = pickle.load(handle)

df_log = pd.DataFrame(logbook)

# plot mean versus generation and best versus generation from generation 20
import matplotlib.pyplot as plt
import numpy as np
df_log = df_log[df_log['gen'] > 10]
plt.plot(df_log['gen'], df_log['avg'], label='mean')
plt.plot(df_log['gen'], df_log['min'], label='best')
plt.legend()
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Mean and best fitness versus generation')
plt.show()
```
